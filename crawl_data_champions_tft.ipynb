```python
# Import requests for crawling a page and Beautiful Soup for parse the HTML content
# Import other libraries for demonstration
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fetch a website URL and store it to webpage
overview_webpage_a = requests.get("https://blitz.gg/tft/champions/overview?set=8&sortBy=name&sortDirection=ASC")
overview_webpage_z = requests.get("https://blitz.gg/tft/champions/overview?set=8&sortBy=name&sortDirection=DESC")
#Take the webpage variable and fetch the content using BeautifulSoup
soup_overview_a = BeautifulSoup(overview_webpage_a.content, "html.parser")
soup_overview_z = BeautifulSoup(overview_webpage_z.content, "html.parser")
```

```python
# Store the whole text in HTML that have the class "Rating" and "CocoaPercent"
name_champions_a = soup_overview_a.find_all(attrs={"class": "type-body2"})   # 0 la ten, 1 la gia, 8 la tam danh
traits_a = soup_overview_a.find_all(attrs={"class": "⚡a40f4298 ⚡904d395"}) # moi phan tu se chua toc he trong name cua 1 vi tuong theo thu tu 1, 3, 5, ...

# Store the whole text in HTML that have the class "Rating" and "CocoaPercent"
name_champions_z = soup_overview_z.find_all(attrs={"class": "type-body2"})   # 0 la ten, 1 la gia, 8 la tam danh
traits_z = soup_overview_z.find_all(attrs={"class": "⚡a40f4298 ⚡904d395"}) # moi phan tu se chua toc he trong name cua 1 vi tuong theo thu tu 1, 3, 5, ...

# Them gia tri bi khuyet trong web
name_champions_z.insert(248, 0)
name_champions_z.insert(249, 0)
```

```python
# print(len(name_champions_a))
# for i in range(0, len(name_champions_a), 9):
#     print(name_champions_a[i].text)

# for i in range(0, len(name_champions_z), 9):
#     print(name_champions_z[i].text)
trait_champions = []
list_name_champions = []
price_champions = []

for i in range(1, len(traits_a), 2):
    pos = int(((i-1)/2)*9)
    if(list_name_champions.count(name_champions_a[pos].text) == 0):
        trait_champions.append([])
        price_champions.append(int(name_champions_a[pos+1].text))
        list_name_champions.append(str(name_champions_a[pos].text))
        for j in traits_a[i].find_all('div', class_ = "⚡afdc06"):
            temp = (j['name'])
            trait_champions[-1].append(temp)

for i in range(1, len(traits_z), 2):
    pos = int(((i-1)/2)*9)
    if(list_name_champions.count(name_champions_z[pos].text) == 0):
        trait_champions.append([])
        price_champions.append(int(name_champions_z[pos+1].text))
        list_name_champions.append(str(name_champions_z[pos].text))
        for j in traits_z[i].find_all('div', class_ = "⚡afdc06"):
            temp = (j['name'])
            trait_champions[-1].append(temp)
        
```

```python
data_champions_tft_df = pd.DataFrame(data = {'Traits': trait_champions, 'Price': price_champions},index = list_name_champions).sort_index()
data_champions_tft_df
```

Stat

```python
# Fetch a website URL and store it to webpage
stat_webpage_a = requests.get("https://blitz.gg/tft/champions/stats?set=8&sortBy=champion&sortDirection=DESC")
stat_webpage_z = requests.get("https://blitz.gg/tft/champions/stats?set=8&sortBy=champion&sortDirection=ASC")

#Take the webpage variable and fetch the content using BeautifulSoup
soup_stat_a = BeautifulSoup(stat_webpage_a.content, "html.parser")
soup_stat_z = BeautifulSoup(stat_webpage_z.content, "html.parser")
```

```python
# Store the whole text in HTML
champion_stat_a = soup_stat_z.find_all(attrs={"class": "type-body2"})    # 0: ten tuong, 1: vi tri TB, 2: top4, 3: ti le win, 4: tong so tran
champion_stat_z = soup_stat_a.find_all(attrs={"class": "type-body2"})    # 0: ten tuong, 1: vi tri TB, 2: top4, 3: ti le win, 4: tong so tran
```

```python
stat_champions_df = pd.DataFrame(index=list_name_champions, columns=['Avg Place', 'Top 4', 'Top 1'])

for i in range(0, len(champion_stat_a), 5):
    name = champion_stat_a[i].text
    stat_champions_df.loc[name]['Avg Place'] = float(champion_stat_a[i+1].text)
    stat_champions_df.loc[name]['Top 4'] = str(champion_stat_a[i+2].text)
    stat_champions_df.loc[name]['Top 1'] = str(champion_stat_a[i+3].text)

for i in range(0, len(champion_stat_z), 5):
    name = champion_stat_z[i].text
    stat_champions_df.loc[name]['Avg Place'] = float(champion_stat_z[i+1].text)
    stat_champions_df.loc[name]['Top 4'] = str(champion_stat_z[i+2].text)
    stat_champions_df.loc[name]['Top 1'] = str(champion_stat_z[i+3].text)
```

```python
data_champions_tft_df = data_champions_tft_df.assign(Avg_Place = stat_champions_df['Avg Place'], Top_4 = stat_champions_df['Top 4'], Top_1 = stat_champions_df['Top 1'])
```
